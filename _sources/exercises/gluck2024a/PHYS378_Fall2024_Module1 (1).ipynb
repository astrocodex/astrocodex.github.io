### Accuracy and Speed

*Read Chapters 3 and 4 of “Computational Physics with Python” by Mark Newman (available online).*

To first understand why taking a look at accuracy and speed are important, let's look at the value of pi that Python uses in its calculations and compare it to the actual value of pi:

$$ \texttt{True value of pi:	3.1415926535897932384626…} \\
\texttt{Value in Python:	3.1415926535897931} \\
\texttt{Difference:		0.0000000000000001384626…} $$

There is a rounding error! Any number whose true value has more than 16 significant figures will get rounded off. Computers cannot be infinitely accurate, but this for the time being is good enough!

**Things to Avoid:**

You should never use an if statement to test the equality of two floats.



````python
x=3.0
if x == 3.3:
  print(x) # this doesn't give any error even though it should!
````

It rarely occurs in physics calculations that you need to test the equality of floats, but if you do, then you should do something like this instead:



````python
x=3.3
epsilon = 1e-12
if abs(x-3.3)<epsilon:
  print(x)
````

#### Mathematical Context



However, this might not do what you want it to do...

An error of one part in $10^{16}$ does not seem too bad. But what happens if we now add, subtract, or otherwise combine several different numbers, each with its own error?

In general if x is accurate to a certain number of significant digits, say 16, then the rounding error will have a typical size of x/$10^{16}$.

$$ c \approx 10^{16} $$

It is usually a good assumption to consider the error to be a (uniformly distributed) random number with standard deviation $\sigma$ where

$$ \sigma = cx $$



1. Addition/Subtraction: If we add or subtract two numbers $x_1$ and $x_2$ with standard deviations $\sigma_1$ and $\sigma_2$, then the error on the sum or difference $x=x_1 \pm x_2$ is given by the **Central Limit Theorem**:
> The variance of $\sigma^2$ of the sum or difference is equal to the sum of the individual variances:
$$ \sigma^2 = \sigma_1^2 + \sigma_2^2$$
Hence, the standard deviation of the sum or difference is,
$$\sigma = \sqrt{\sigma_1^2 + \sigma_2^2}$$
So the error if we add or subtract two numbers ($x_1$ and $x_2$), their total standard deviation becomes,
$$\sigma = c\sqrt{\sigma_1^2 + \sigma_2^2}$$

2. Similarly if we multiply or divide two numbers then the variance of the result x obeys:
$$\frac{\sigma^2}{x^2} = \frac{\sigma_1^2}{x_1^2} + \frac{\sigma_2^2}{x_2^2}$$
Such that we obtain $\sigma = \sqrt{2}cx$.

The value of $\sigma$ increases in size as the more numbers we combine, N, increases, including the larger the error on the results. Although the increase is a relatively slow one, proportional to the square root of N.

We can also ask about the fractional error on $\Sigma_i x_i$, i.e., the total error divided by the value of the sum. The size of the fractional error is given by,

$$\frac{\sigma}{\Sigma_i x_i} = \frac{c\sqrt{N \bar{x^2}}}{Nx} = \frac{c\sqrt{\bar{x^2}}}{x\sqrt{N}}$$

where $\bar{x}$ is the mean value of x. **The fractional error in the sum actually goes down as we add more numbers.**

At the first glance, this appears to be pretty good. So, what’s the problem? Actually, there are a couple of them:

1. When the size of the numbers you’re adding vary wildly, then the smaller numbers may get lost.

2. The most severe problems arise when you are not adding but subtracting numbers!

**Exercise:** Quantum harmonic oscillators at a finite temperature (ignore the complex terminology for now, just take a look at the equations!)

We can describe the energy levels of a quantum harmonic oscillator by,

$$E_n = \hbar \omega \left( n + \frac{1}{2}\right)$$

where n=0,1,2,...

We can describe the average energy of this oscillator at some temperature, T, by

$$\langle E \rangle = \frac{1}{Z} \sum_{n=0}^{\infty} E_n e^{-\beta E_n}$$

where $\beta = 1/(k_B T)$, with $k_B$ as the Boltzmann constant and

$$Z = \sum_{n=0}^{\infty} e^{-\beta E_n}$$.


> <font color="skyblue"> **Your task:** Find $\langle E \rangle$ at $k_B T=100$. Use $\hbar=\omega=1$ for this calculation. *Notice that the terms in the sum for* $\langle E \rangle$ *and* $Z$ *decrease as n increases. Therefore, we can approximate these infinite sums by taking the first 1000 terms.* </font>

````python
# Your work here!
````

````python
terms = 1000
beta = 1.0/100
S = 0.0
Z = 0.0
for n in range (terms):
	E = n + 0.5
	weight = exp(-beta*E)
	S += weight*E
	Z += weight

print(S/Z)
````
